```python
import math
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook


n=1000
fig=plt.figure()
ax=plt.axes(projection='3d')


theta_max=20*np.pi
theta=np.linspace(0,theta_max,n)
a=5
b=7
x=a*np.sin(theta)
y=b*np.cos(theta)
z =theta
ax.plot(x,y,z,'b',lw=2)

theta_current = 3*np.pi/2

x_0=a*math.cos(theta_current)
y_0=b*math.sin(theta_current)
z_0=theta_current

x_1=a*math.cos(theta_current)
y_1=b*-math.sin(theta_current)
z_1=1



x_2=x_0+x_1
y_2=y_0+y_1
z_2=z_0+z_1

x_s=[x_0,x_2]
y_s=[y_0,y_2]
z_s=[z_0,z_2]

ax.plot(x_s,y_s,z_s)

plt.show()
```
